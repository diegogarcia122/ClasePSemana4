```python
import pandas as pd

df = pd.read_csv('train.csv')

print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
product_freq = df['Product Name'].value_counts()
print(product_freq.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_5_stores = df.groupby('Customer Name')['Sales'].sum().nlargest(5)
top_5_stores.plot(kind='bar', title='Top 5 Clientes con Mayores Compras')
plt.ylabel('Compras Totales')
plt.show()
```

```python
plt.hist(df['Category'], bins=20, color='skyblue')
plt.title('Distribucin de las Categorias')
plt.xlabel('Categorias en Ventas')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.scatter(df['Ship Mode'], df['Sales'], alpha=0.5)
plt.title('Relacin entre Tipo de Envo y Ventas')
plt.xlabel('Tipo de Envo')
plt.ylabel('Cantidad de Ventas')
plt.ylim([0, 10])
plt.show()
```
